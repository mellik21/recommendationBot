```python
import pandas as pd
import numpy as np

data = pd.read_csv('cleaned_anime.csv')
```

```python
data.head()
```

```python
unic_genres = []
for genres in data.Genres:
        genres = genres.replace('[','').replace(']','').replace('\'','').split(', ')

        for genre in genres:
            if genre not in unic_genres:
                unic_genres.append(genre)
```

```python
genres = pd.DataFrame()
unic_genres = sorted(unic_genres)
genres['name'] = unic_genres

genres
```

```python
genres.to_csv('genres.csv', index=True, encoding='utf-8')
```

```python
# Заменяем все жанры на их id
for index, row in data.iterrows():
    g = row['Genres'].replace('[','').replace(']','').replace('\'','').split(', ')
    new_genres = []
    for genre in g:
        new_genres.append(genres.index[genres['name']==genre].tolist()[0])
     
    data['Genres'][index] = new_genres
```

```python
data.head()

data = data.fillna(-1)
```

```python
data.to_csv('not_none_cleaned_data.csv', index=True, encoding='utf-8')
```
